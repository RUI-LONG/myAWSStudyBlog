```python
import awswrangler as wr
import pandas as pd
import boto3
```

```python
region = boto3.session.Session().region_name
bucket = f'beyoung-{region}-20200823'
prefix = 'geo-spatial-tutorial'
```

```python
fp_acc = f's3://{bucket}/{prefix}/accommodations.csv'
fp_zip = f's3://{bucket}/{prefix}/zipcode.csv'
# DE.txt
```

```python
df_acc = wr.s3.read_csv(fp_acc, sep=';')
df_acc.head()
```

```python
df_zip = wr.s3.read_csv(fp_zip, sep=';')
df_zip.head()
```

## Writing Parquet files

```python
fp_acc_parquet = f"s3://{bucket}/{prefix}/parquet/accommodations.parquet"
fp_zip_parquet = f"s3://{bucket}/{prefix}/parquet/zipcode.parquet"

wr.s3.to_parquet(df_acc, fp_acc_parquet)
wr.s3.to_parquet(df_zip, fp_zip_parquet);
```

```python
## Reading single Parquet file
wr.s3.read_parquet(fp_acc_parquet)
```

# Checking Glue Catalog Databases

```python
dbname = 'accommodations-20200923'

databases = wr.catalog.databases()
if dbname not in databases.values:
    wr.catalog.create_database(dbname)
    print(wr.catalog.databases())
else:
    print(f"Database {dbname} already exists")
```

## Writing DataFrames to S3 Parquet and Glue Catalog

```python
fp_glue_db = f's3://{bucket}/{prefix}/airbnb/'
```

```python
desc = "This is my accommodations table."

param = {
    "source": "airbnb",
    "class": "e-commerce"
}

res = wr.s3.to_parquet(
    df=df_acc,
    path=fp_glue_db,
    dataset=True,
    database=dbname,
    table="accommodations",
    mode="overwrite",
    description=desc,
    parameters=param
)
```

```python
wr.catalog.table(database=dbname, table="accommodations")
```

```python
fp_glue_db_zip = f's3://{bucket}/{prefix}/zipcode/'
```

```python
desc = "This is my zipcode table."

param = {
    "source": "EU-DE",
    "class": "e-commerce"
}

res = wr.s3.to_parquet(
    df=df_zip,
    path=fp_glue_db_zip,
    dataset=True,
    database=dbname,
    table="zipcode",
    mode="overwrite",
    description=desc,
    parameters=param
)
```

```python
wr.catalog.table(database=dbname, table="zipcode")
```
